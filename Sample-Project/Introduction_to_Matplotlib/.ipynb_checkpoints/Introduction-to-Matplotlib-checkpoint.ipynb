### This is Object-Oriented API

But in Object-Oriented API, there are lot of methods to explore. So lets explore them all...

```python
# 1st method
fig  = plt.figure() #creates a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() #creates a figure
ax = fig.add_axes([1, 1, 1, 1]) #add some axes
ax.plot(x, y)
plt.show();
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
# ax.plot(x, y);
ax.plot(x, [50, 100, 250, 900]) # add some plot
type(fig), type(ax)
```

### Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data 
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot",
      xlabel="x-label",
      ylabel="y-label")

# 5. Save & show (You will save the whole figure)
fig.savefig("../data_files/images/sample-plot.png")
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2,
                                            figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot Option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should You use? (pyplot vs matplotlib ObjectOriented method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method 

```python
heart_disease.head()
```

```python
# patients over 50 ages.....
over_50 = heart_disease[heart_disease["age"] > 50]
print(len(over_50))
over_50.head()
```

```python
# Scatter plot of age and cholastrol with color according the targetS
# Using PyPlot method 
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            figsize=(10, 6));
```

```python
# MatplotLib OO method mixed eith pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# Now the set_xlim() is used to extend the range/limits of x axis
# in previous figure the x limit was 50 to 80
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);

ax.set_xlim([45, 100]);
```

 ### Customization in figure using OO method

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data using Scatter Plot
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend method
ax.legend(*scatter.legend_elements(), title="Target");

# Add a Horizontal Line
ax.axhline(over_50["chol"].mean(),
                    linestyle='--');
```

```python
# Full Power of the OO Method
# Adding subplots of chol, age thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                               ncols=1,
                              figsize=(10, 15),
                              sharex=True)

# Add data to ax0
scatter_ax0 = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a Mean Line in the ax0
ax0.axhline(y=over_50["chol"].mean(), linestyle=":");



# Now add data to ax1
scatter_ax1 = ax1.scatter(x=over_50["age"],
                         y=over_50["thalach"],
                         c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a Legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")

#Add a Mean Line in ax1
ax1.axhline(y=over_50["thalach"].mean(), linestyle="-.")


# Add a title to Figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")


fig.savefig("../data_files/images/heart-disease-fig-using-dataframe.png");
```

```python
# changing colors in styles......


# Set a style
plt.style.use('seaborn-whitegrid')

# Create a Figure and axes
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data and adding colors...
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"], 
                    cmap="winter") # this changes the color scheme

# add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--',
          color='brown');
```

```python
plt.style.available
```

```python
# Customizing the y and x axis limitations
plt.style.use('seaborn-talk')

# Subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# add data to ax0 and ax1
scatter_ax0 = ax0.scatter(x=over_50["age"],
                        y=over_50["chol"],
                        c=over_50["target"],
                        cmap="winter")

scatter_ax1 = ax1.scatter(x=over_50["age"],
                         y=over_50["thalach"],
                         c=over_50["target"],
                         cmap="winter")

# add set() to ax0 and ax1
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")

ax1.set(title="Heart Disease and Heart Rate Moniter",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Changing the limits of x-axis
ax1.set_xlim([50, 80]) # we only need to set one the x axis of 2 axises because they are sharing each other

# Changing the limits of y-axis
ax0.set_ylim([100, 450])
ax1.set_ylim([80, 200])


# add a legend to ax0 and ax1
ax0.legend(*scatter.legend_elements(), title="Target")
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a MeanLine
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--",
           color="brown")
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--",
           color="brown")

# Add super title to Figure
fig.suptitle("Heart Disease Analysis",
            fontsize="16",
            fontweight="bold")

# Saving Figure
fig.savefig("../data_files/images/heart-disease-analysis-styled.png");
```
